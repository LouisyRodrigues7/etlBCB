```python

estatisticas = df.describe().T # Método Pandas para criar um resumo estatístico das colunas numericas do Dataframe e df para calcular media, desvio padrao... 
estatisticas['coef_var'] = estatisticas['std'] / estatisticas['mean']  # std = pega linha que cotem os valores do desvio padrao para cada coluna 
estatisticas # mean = pega linha que contem as médias para cada coluna
```

```python
import matplotlib.pyplot as plt #biblioteca de visualização de dados em Python

# Gráfico de barras das médias
estatisticas['mean'].sort_values(ascending=False).plot(kind='bar', figsize=(14,6), title='Média das variáveis numéricas por trimestre') # sort_values ordenar valores da coluna e o ascending=false decrescente
plt.ylabel('Média (valores monetários)') # descreve o que o eixo Y representa (média dos valores monetários )
plt.tight_layout() #Esse método ajusta automaticamente o layout do gráfico
plt.grid(True) #Esse comando adiciona uma grade (linhas horizontais e verticais) ao gráfico
plt.show() #Exibe o grafico
```

```python
estatisticas.sort_values('mean', ascending=False) # .sort_values() é utilizado para ordenar as linhas do DataFrame com base nos valores de uma coluna específica. (média) 
#ascending=False define a ordem decrescente 
```

```python
# Quais variáveis têm uma variação mais significativa em relação à sua média, podendo ser mais imprevisíveis.
maior_desvio_padrao = estatisticas['std'].sort_values(ascending=False)
print(" Variáveis com maior desvio padrão:")
print(maior_desvio_padrao)


maior_coef_var = estatisticas['coef_var'].sort_values(ascending=False)
print("Variáveis com maior coeficiente de variação (mais voláteis):")
print(maior_coef_var)
```

## Estudo da mediana do valorPix

Para encontrar a mediana precisamos ordenar os valores e encontrar os elementos centrais.

Se o número de elementos for par, usamos a fórmula:


$$ \frac{X_\frac{n}{2} + X_(\frac{n}{2} +1)}{2} $$


para determinar o elemento


```python
ordenados = df['valorPix'].sort_values()
n = len(ordenados)

print("A lista possui: ", n," elementos")
```

Para determinar a mediana devemos encontrar os números que estão na posição 12 e 13, mas as listas em python começam a contar da posição 0. Logo a mediana é o valor médio entre os elementos na posição 11 e 12.

```python
inf = ordenados [n/2 -1]
sup = ordenados[n/2]

mediana = (inf + sup)/2

mediana
```

Para o caso de um número de elementos ímpar a mediana é o valor central:

$$ \frac{X_{n+1}}{2}$$

## Cálculo da mediana no Pandas

```python
df['valorPix'].median()
```

## Medidas de Dispersão 

### Variância 

A média das diferença quadrada entre os valores observados e o valor médio do conjunto de dados. Isso também é conhecido como dispersão média. 

$$ \frac{|X_1-\bar{X}|2+...+|X_n-\bar{x}|^2}{n}$$



```python
df ['valorPix'].var()
```

obs.: A variância apresenta valores altos e a unidade de medida é quadrada

### Desvio-Padrão   

É a raiz quadrada da variância

```python
df['valorPix'].std()
```

obs.: Em muitos casos 66% dos dados estão a um desvio-padrão da média do seu conjunto de dados.
